## Count Positive, Negative, and Zero Elements

### Subtask:
Write a Python program using NumPy to count and display the number of positive, negative, and zero elements present in an array.


**Reasoning**:
The subtask requires counting positive, negative, and zero elements in a NumPy array. I will use boolean indexing and the `sum()` method to efficiently count these categories and then display the results with clear labels.



````python
import numpy as np

# 1. Create a NumPy array with mixed values
my_array = np.array([-5, 0, 10, -2, 7, 0, 1, 15, -8, 0])

# 2. Count positive elements
positive_count = (my_array > 0).sum()

# 3. Count negative elements
negative_count = (my_array < 0).sum()

# 4. Count zero elements
zero_count = (my_array == 0).sum()

# 5. Print the results
print("Original Array:", my_array)
print("Number of Positive Elements:", positive_count)
print("Number of Negative Elements:", negative_count)
print("Number of Zero Elements:", zero_count)
````

## Bubble Sort Algorithm

### Subtask:
Implement the bubble sort algorithm in Python to sort the elements of an array.


**Reasoning**:
The subtask requires implementing the bubble sort algorithm to sort a NumPy array. I will define the `bubble_sort` function, apply the sorting logic with nested loops and swapping, create an example array, call the function, and then print the original and sorted arrays to demonstrate the result.



````python
import numpy as np

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Create an example NumPy array
example_array = np.array([64, 34, 25, 12, 22, 11, 90])

# Make a copy to preserve the original for printing
original_array = np.copy(example_array)

# Call the bubble_sort function
sorted_array = bubble_sort(example_array)

# Print the original and sorted arrays
print("Original Array:", original_array)
print("Sorted Array (using Bubble Sort):", sorted_array)
````

## Merge Two Arrays Implementation

### Subtask:
Create two example 1D NumPy arrays and use `np.concatenate()` to merge them into a single new array. Then, print all three arrays to verify the merging operation.


**Reasoning**:
To merge two 1D NumPy arrays, I will first define two example arrays. Then, I will use the `np.concatenate()` function to combine them into a single array. Finally, I will print all three arrays (the two original ones and the merged one) to confirm that the operation was successful and as expected.

**Reasoning**:
I will create two 1D NumPy arrays, concatenate them using `np.concatenate()`, and then print all three arrays (the two original and the merged one) to fulfill the subtask requirements.



````python
import numpy as np

# 1. Create two 1D NumPy arrays
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

# 2. Use np.concatenate() to merge array1 and array2
merged_array = np.concatenate((array1, array2))

# 3. Print array1, array2, and merged_array
print("Array 1:", array1)
print("Array 2:", array2)
print("Merged Array:", merged_array)
````
